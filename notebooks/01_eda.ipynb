```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Carregar os dados

```python
train = pd.read_csv('../data/raw/train.csv')
```

# Visão geral do conjunto de dados

```python
print(train.shape)
print(train.info())
train.describe().T
```

## Distribuição da variável alvo: SalePrice

```python
sns.histplot(train['SalePrice'], kde=True)
plt.title('Distribuição de SalePrice')
plt.xlabel('SalePrice')
plt.ylabel('Frequência')
plt.show()
```

# Correlação entre variáveis numéricas

```python
correlation_matrix = train.corr(numeric_only=True)
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm', vmax=0.8)
plt.title('Mapa de Correlação entre Variáveis Numéricas')
plt.show()
```

## Principais correlações com SalePrice

```python
top_corr = correlation_matrix['SalePrice'].abs().sort_values(ascending=False).head(10)
print(top_corr)
```

## Relação entre variáveis significativas e SalePrice

```python
sns.scatterplot(data=train, x='GrLivArea', y='SalePrice')
plt.title('GrLivArea vs SalePrice')
plt.show()

sns.boxplot(x='OverallQual', y='SalePrice', data=train)
plt.title('OverallQual vs SalePrice')
plt.show()
```
